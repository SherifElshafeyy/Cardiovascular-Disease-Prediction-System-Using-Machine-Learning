```python
import pandas as pd
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
education_dict = {1:"High School",2:"High School Diploma/GED",3 : "College" , 4:"Higher Degree"}
education_dict
```

```python
education_dict[1]
```

```python
data['education'] = data['education'].map(education_dict)
```

```python
data.rename(columns={"male":"gender", "TenYearCHD":"target"}, inplace=True)
```

```python
gender_dict = {1:"Male" , 0 : "Female"}
```

```python
data['gender'] = data['gender'].map(gender_dict)
```

```python
data.to_csv('data_v1.csv')
```
