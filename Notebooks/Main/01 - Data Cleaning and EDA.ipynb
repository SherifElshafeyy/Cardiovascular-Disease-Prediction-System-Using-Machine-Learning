### Detected Problems

* Nulls Values


```python
def col_details(col_name):
    
    print(f"####### {col_name} ########")
    
    print(f"\nNumber of unique Values {data[col_name].nunique()}")
    
    print(f"There are {data[col_name].isnull().sum()} Null Values")
    random_number = np.random.randint(4000)
    print(f"\nSome Random Data \n {data[col_name][random_number:random_number + 5].values}")
    print(f"\nTop 5 Values {data[col_name].value_counts().index[:5]}")
    
    print(f"Max value is {data[col_name].max()} Min {data[col_name].min()}")
```

### Male

```python
col_details("male")
```

### age

```python
col_details("age")
```

### education
* Null values : there are many ways to handle null values such as replacing them with Medain or Mean or using some algorthim for numerical data

```python
col_details("education")
```

```python
print(f"{data['education'].isnull().sum() / len(data) * 100 } %")
```
